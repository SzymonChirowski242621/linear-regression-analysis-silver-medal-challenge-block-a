# Data Exploration to perform Linear Regression on the SDG dataset

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

## Load Data

```python
df = pd.read_excel('data/processed_data.xlsx')
df.head()
```

## Data Exploration

```python
df.info()
```

```python
df.describe()
```

## Data Visualization

### Scatter Plot to visualize the relationship between GDP and other variables

```python
plt.figure(figsize=(10, 6))
plt.title('GDP vs P_KM')
sns.scatterplot(data=df, x='normalized_PKM', y='normalized_GDP')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('GDP vs T_KM')
sns.scatterplot(data=df, x='normalized_TKM', y='normalized_GDP')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('GDP vs TEU')
sns.scatterplot(data=df, x='normalized_TEU', y='normalized_GDP')
plt.show()
```

### Distribution Plot to visualize the distribution of the variables

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of P_KM')
sns.histplot(data=df, x='normalized_PKM', bins=30, kde=True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of T_KM')
sns.histplot(data=df, x='normalized_TKM', bins=30, kde=True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of TEU')
sns.histplot(data=df, x='normalized_TEU', bins=30, kde=True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of GDP')
sns.histplot(data=df, x='normalized_GDP', bins=30, kde=True)
plt.show()
```

## Calculate Correlation

```python
df[['normalized_PKM', 'normalized_TKM', 'normalized_TEU', 'normalized_GDP']].corr()
```

### Plot Correlation Matrix

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[['normalized_PKM', 'normalized_TKM', 'normalized_TEU', 'normalized_GDP']].corr(), annot=True, cmap='coolwarm')
plt.show()
```
