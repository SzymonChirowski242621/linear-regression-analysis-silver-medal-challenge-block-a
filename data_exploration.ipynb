# Data Exploration to perform Linear Regression on the SDG dataset

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

## Load Data

```python
df = pd.read_excel('data/pivot_merged_table.xlsx')
df.head()
```

## Data Exploration

```python
df.info()
```

```python
df.describe()
```

```python
df['Year'].value_counts()
```

## Data Visualization

```python
df['Year'].value_counts().plot(kind='bar', color='skyblue')
```

### Scatter Plot to visualize the relationship between GDP and other variables

```python
plt.figure(figsize=(10, 6))
plt.title('GDP vs P_KM')
sns.scatterplot(data=df, x='P_KM', y='GDP')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('GDP vs T_KM')
sns.scatterplot(data=df, x='T_KM', y='GDP')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('GDP vs TEU')
sns.scatterplot(data=df, x='TEU', y='GDP')
plt.show()
```

### Distribution Plot to visualize the distribution of the variables

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of P_KM')
sns.histplot(data=df, x='P_KM', bins=30, kde=True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of T_KM')
sns.histplot(data=df, x='T_KM', bins=30, kde=True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of TEU')
sns.histplot(data=df, x='TEU', bins=30, kde=True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of GDP')
sns.histplot(data=df, x='GDP', bins=30, kde=True)
plt.show()
```

## Calculate Correlation

```python
df[['P_KM', 'T_KM', 'TEU', 'GDP']].corr()
```

### Plot Correlation Matrix

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[['P_KM', 'T_KM', 'TEU', 'GDP']].corr(), annot=True, cmap='coolwarm')
plt.show()
```
